```python
import easyocr
import cv2
import re

class TamakkanOCR:
    def __init__(self, use_gpu=True):
        self.reader = easyocr.Reader(['en'], gpu=use_gpu)

    def read_sign(self, cropped_frame):
        if cropped_frame is None or cropped_frame.size == 0:
            return ""

        results = self.reader.readtext(cropped_frame)
        detected_text = " ".join([res[1] for res in results])
        detected_text = detected_text.strip().upper()
        detected_text = re.sub(r'[^A-Z0-9 ]', '', detected_text)
        return detected_text
```
